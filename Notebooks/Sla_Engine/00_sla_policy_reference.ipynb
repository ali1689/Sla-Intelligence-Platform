```python
# Notebook: 00_ref_sla_policy
# Layer   : Reference
# Purpose : Define SLA response & resolution targets
# Source  : Client SLA agreement
# ============================================================
from pyspark.sql import SparkSession
# -----------------------------

# -----------------------------
# 2. Define SLA policy data
# -----------------------------
sla_data = [
   # priority, response_minutes, resolution_hours, impact, emergency, notes
   ("P1", 15, 4, "Blocking", "High", "Critical incident"),
   ("P2", 30, 8, "Blocking/Major", "High", "High priority incident"),
   ("P3", 60, 12, "Major/Minor", "Medium", "Moderate incident"),
   ("P4", 1440, 72, "Minor", "Low", "Low priority incident"),
]
columns = [
   "priority",
   "response_time_minutes",
   "resolution_time_hours",
   "impact_level",
   "emergency_level",
   "description"
]
sla_df = spark.createDataFrame(sla_data, columns)
# -----------------------------
# 3. Persist as Delta table
# -----------------------------
spark.sql("DROP TABLE IF EXISTS slainte_bronze.sla_policy")
sla_df.write \
   .format("delta") \
   .mode("overwrite") \
   .saveAsTable("slainte_bronze.sla_policy")
# -----------------------------
# 4. Validation
# -----------------------------
spark.table("slainte_bronze.sla_policy").show(truncate=False)
```
